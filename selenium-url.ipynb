```python
from selenium import webdriver
import requests
import sys, getopt
import argparse
from webdriver_manager.chrome import ChromeDriverManager

chrome_options = webdriver.ChromeOptions()
chrome_options.add_argument("--headless")
driver = webdriver.Chrome(ChromeDriverManager().install(), chrome_options=chrome_options)
full_text=[]

def parse_args():
    parser = argparse.ArgumentParser()
    parser.add_argument('-i', '--infile', default='', help='input filename')
    parser.add_argument('-o', '--outfile', default='', help='output filename')
    return parser.parse_args()

def main():
    args = parse_args()
    outfile = args.outfile
    infile = args.infile
    
    with open(infile) as f:
        content = f.read().splitlines()
    f.close()
    
    f = open(outfile, "w")
    for u in content:
        driver.get(u)
        elems = driver.find_element_by_tag_name('body').text
        full_text.append(elems)
    print(full_text)
    
    f.write(str(full_text))
    
    # print(full_text)
    driver.close()
    f.close()
main()
```

```python
from collections import Counter # ntuk menghitung jumlah kemunculan setiap elemen pada sebuah iterable (seperti list, tuple, atau string)
import re  # berfungsi untuk memanipulasi string menggunakan ekspresi reguler
# regular expressions

artikel = article.text 
artikel = artikel.lower() # agar tidak ada huruf kapital

# menggunakan fungsi re.sub() untuk subtitusi var string
artikel_bersih = re.sub(r"[^a-zA-Z\s]", "", artikel) 
bagi = artikel.split() # membuatnya menjadi list
frekuensi = Counter(bagi) # memberikan seberapa sering kata itu muncul


print('List dan kata yang muncul lebih dari 5 kali akan tersimpan dalam bentuk .txt')

with open("KaesangP.txt", "w") as file: # menyimpan hasil output dlm bentuk .txt
    for kalimat in artikel_bersih: # menggunakan looping for
        file.write(f"{artikel_bersih}") # menyimpan hasil cleaning 
        file.write(f"{frekuensi}\n") # menyimpan hasil dari pembagian per kata dalam list 
        break
    for kata,frekuensi in frekuensi.items():
        if frekuensi >= 5 : # apabila kata dalam list muncul lebih dari sama dengan 5 kali maka
            file.write(f"{kata}:{frekuensi} kali \n") # akan tersimpan dalam file .txt dan akan memunclkan berapa kali kata tersebut muncul
```
